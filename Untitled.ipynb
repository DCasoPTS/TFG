```python
import sys
from keras import models
import pandas as pd
import random
import pandas as pd
import numpy as np
from keras import layers
from keras import models
from keras import backend as K
from keras import callbacks
from keras import optimizers
from keras.regularizers import l2
# from keras.utils import to_categorical
import matplotlib.pyplot as plt
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from keras.layers import Dense, Dropout, Flatten
```

```python
data = pd.read_csv('/Users/diego/venv/data.csv', delimiter='|')
data = data.sample(frac=1).reset_index(drop=True)
train = data[:1179509]
test = data[1179509:]
print(data.shape[0], train.shape[0], test.shape[0])
```

```python
variables = ['Lepton_pdgId[0]', 'Lepton_pdgId[1]', 'Lepton_pt[0]', 'Lepton_pt[1]', 'mll', 'ptll', 'mth', 'mtw1', 'mtw2', 'PuppiMET_pt', 'TkMET_pt', 'mpmet', 'drll', 'dphill', 'dphilmet', 'dphilmet1', 'dphilmet2']
genpT = ["type"]
```

```python
model = models.Sequential()
model.add(layers.Dense(256, activation='relu',input_dim=len(variables)))

for i in range(0, 10):
    if i == 0:
        currentNeurons = 256

    if random.choice([True, False]):
        model.add(layers.Dense(currentNeurons, activation='relu'))
    elif currentNeurons == 2:
        model.add(layers.Dense(currentNeurons, activation='relu'))
    else:
        currentNeurons = currentNeurons/2
        model.add(layers.Dense(currentNeurons, activation='relu'))
#     model.add(Dropout(0.2))
model.add(layers.Dense(1, activation='sigmoid'))
```

```python
epochs = 200
# lr = 0.00044444422525
lr = 0.0000001
decay = lr/epochs
opt = optimizers.Adam(lr, decay)
model.compile(loss="binary_crossentropy", optimizer=opt, metrics=['accuracy'])

history = model.fit(train[variables],train[genpT],shuffle=True,validation_split=0.1,\
        epochs=epochs, batch_size=512, verbose=2, callbacks=[callbacks.EarlyStopping(monitor='val_loss',\
        patience=20,verbose=1)])
# model.save('model_nFirstNeurons' + 256 + '_nHiddenLayers_' + 10 + '_LearningRate_'\
#  + lr + '_BatchSize_' + 512 + '.h5')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper right')
#plt.legend(['train'], loc='upper right')
plt.show()
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='lower right')
#plt.legend(['train'], loc='lower right')
plt.show()
```
